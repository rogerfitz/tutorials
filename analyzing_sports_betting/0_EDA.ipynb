```python
altair_df=df.copy()

altair_df=altair_df.reset_index().rename(columns={"index": "year"})
altair_df=pd.melt(altair_df,id_vars=["year"],value_vars=df.columns).sort_values(by=["year","variable"])
altair_df
```

```python
highlight = alt.selection_point(on='mouseover',
                          fields=['variable'], nearest=True, bind="legend")
base=alt.Chart(altair_df[altair_df['variable'].isin(["Baseball","Basketball","Football","Total"])]).encode(
    x='year:N',
    y='value:Q',
    color='variable:N',
    opacity=alt.condition(highlight, alt.value(1), alt.value(0.2))
)

points = base.mark_circle().encode(
    opacity=alt.value(0)
).add_params(
    highlight
).properties(
    width=600,
    title="Casino Win Percent by Sport"
)

lines = base.mark_line().encode(
    size=alt.condition(~highlight, alt.value(1), alt.value(3))
)

points + lines
```

# An Analysis of Odds History
Since I feel shakiest about that 3rd claim let's analyze another possible reason. How did the fee charged by sports books change over time. It's hard to find actual payout info so instead we will use the futures market for super bowl odds as a proxy.

```python
#https://www.pro-football-reference.com/years/2005/preseason_odds.htm
import requests
import tqdm
import time
year=1992
req_dict={}
for year in tqdm.tqdm(range(1992, 2022)):
    time.sleep(2)
    req=requests.get(f"https://www.pro-football-reference.com/years/{year}/preseason_odds.htm")
    req_dict[year]=req
```

```python
from bs4 import BeautifulSoup
all_dfs=[]
for year, req in req_dict.items():
    table=BeautifulSoup(req.text).find(id="preseason_odds")
    rows=[]
    columns=[x.text for x in table.find("thead").find_all("th")]
    for row in  table.find("tbody").find_all("tr"):
        if row:
            rows.append([x.text.replace("+","") for x in row.find_all("td")])
    df=pd.DataFrame(rows,columns=columns[1:])
    df['year']=year
    all_dfs.append(df)
odds_df=pd.concat(all_dfs)
odds_df
```

```python
odds_df['implied']=1/(odds_df['Super Bowl Odds'].astype(float)/100+1)
odds_df.groupby("year")[['implied']].sum().plot()
```

Implied odds are the odds you need in order to profit on a bet. For +200 this is a payout of 300 on a bet of 100 so you need to win 1/3 times. The formula is as above. 1/(Payout/100+1)=1/(200/100+1)=1/(2+1)=1/3.
  
So betting odds get better for consumers but the casinos still too home more cash!

Parlays have remained relatively profitable. I tink this can be exmplained by the lottery nature of them. It's well known that people make negative expected value decisions. The lottery is a big example of this. One common explanation of this is the "utility" curve, or how people derive value. When you feel hopeless, you are more prone to taking big chances by focusing on the outcome and not the likelihood. This explains why wealthier people avoid the lottery and poorer people do. Parlays are the same. You have 10 dollars to bet and want to feel a big win, enter a parlay bet where you combine events/legs to win $10,000! That is what people do. Does this mean parlay betting is dumb? No, just on average that parlay betting is dumb.  
  
We don't have much data to support this. Only that parlay betting is pretty consistent for casinos suggesting it's not governed by the same factors as the above. Therefore, sharp bettors are likely an extremely low part of Parlay betting. 

# Takeaway
In a business setting this is hard to explain. You'd really want data on individual bettors, the "to win" amounts, the bet size placed and how that balance came about. An argument towards "dumb" money would show mostly consequential amounts, trying to win back after a losing streak, or a to win amount clustered around major values like 10k, 1k, 500 etc. Betting all that was left in the account is another good indicator.  
  
I'd argue most people listening to your data science report would either come up with arguments like mine or send you on a data digging expedition like we did with the implied odds exercise. At the end of the day, it's explanation and common sense that will sway most decisions. Not fancy statistical tests. And much of the analytics you're asked to do will be confirmatory, aka testing already assumed hypotheses and accepting the first reasonable justification of them.   
  
This is true for most companies that are newly rolling out data science. Once the business wants to get beyond an 80/20 understanding, and you have enough data to do so, a more scientific approach will be used. But even still it can end up being very biased. Just a fancier form of bias.

# More Thoughts
This may feel nihilistic but data science is tough. The best analysis attempts to understand bias and follows the chain of inference closely. Having a hypothesis, then verifying with data simply is great. The more complicated things get and the more hypotheses you develop after seeing data, the more biased you are even if you think you are being smart. Data snooping is a major, major problem. That's why some famous decision makers wait until all the info is available and then quickly make a decision. As opposed to iterating on ideas as info becomes available. That makes it hard to unwind your assumptions and come up with the most rational decision.  

  
I saw this a lot with sports betting. Narrative lockin, X team is a great run d. Once that assumption breaks (injury, few too many flukes) it's hard to realize where you used it. Like in deciding another team is great at running the ball and exucsing another team for having just an OK run D, when in reality their run D is quite suspect.   
  
A more expensive problem is economics. The dataset is so limited, so personal, and so over analyzed. Very prone to data snooping which leads to a more religious approach than a flexible approach.
