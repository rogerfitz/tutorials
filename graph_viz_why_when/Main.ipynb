```python
headers=[th.text for th in soup.find("thead").find_all("th")]
headers[0]="#"
values=[]
for tr in soup.find("tbody").find_all("tr"):
    row=[]
    for td in tr.find_all("td"):
        row.append(td.text)
    if tr.find("th"):
        row.insert(2,tr.find("th").text)
    else:
        continue
    values.append(row)
values[0]
```

```python
players=pd.DataFrame(values, columns=headers)
players['Club']=players['Club'].apply(lambda x: x[1:])
players[["#","Pos.","Player","Goals","Club"]]
```

```python
import numpy as np
import itertools
```

```python
N=len(players)
adj_matrix=np.zeros((N,N))
for club,club_players in players.groupby("Club"):
    for pairs in itertools.combinations(club_players.index,2):
        adj_matrix[pairs[0]][pairs[1]]=1
        adj_matrix[pairs[1]][pairs[0]]=1
G=nx.from_numpy_matrix(adj_matrix)
nx.draw(G,pos=nx.random_layout(G))
```

```python
#to Cytoscape
nodes=[]
edges=[]
for club,club_players in players.groupby("Club"):
    for idx, p in club_players.iterrows():
        nodes.append({"data": {"id": p["Player"],"related":"|".join(club_players['Player'].unique())}})
    for pairs in itertools.combinations(club_players.index,2):
        edges.append({"data": {"source": players.ix[pairs[0]]['Player'], "target": players.ix[pairs[1]]['Player']}})
        edges.append({"data": {"source": players.ix[pairs[1]]['Player'], "target": players.ix[pairs[0]]['Player']}})
elements={"nodes": nodes, "edges":edges}
```

### Hierarchical with Clubs

```python
N=len(players)
club_lookup=players['Club'].unique()
M=len(club_lookup)
adj_matrix=np.zeros((N+M,N+M))
for club,club_players in players.groupby("Club"):
    club_idx=np.where(club_lookup==club)[0][0]+N
    for idx in club_players.index:
        adj_matrix[idx][club_idx]=1
        adj_matrix[club_idx][idx]=1
G=nx.from_numpy_matrix(adj_matrix)
nx.draw(G,pos=nx.spring_layout(G))
```

```python
#to Cytoscape
nodes=[]
edges=[]

for club,club_players in players.groupby("Club"):
    for idx, p in club_players.iterrows():
        nodes.append({"data": {"id": p["Player"],"related":club}})
    nodes.append({"data":{"id": club}})
    club_idx=np.where(club_lookup==club)[0][0]+N
    for idx in club_players.index:
        edges.append({"data": {"target": players.ix[idx]['Player'], "source": club}})
elements={"nodes": nodes, "edges":edges}
```
